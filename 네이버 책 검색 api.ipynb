```python
import requests

# ✅ 1. 네이버 개발자센터에서 발급받은 클라이언트 ID와 시크릿 입력
# 🎯 여기에 본인의 값 입력
client_id = '8xWBKSelQiATYZJCce6v'         # 예: 'abcd1234abcd1234abcd'
client_secret = 'XUcKjn3UPc' # 예: 'efgh5678efgh5678efgh'

# ✅ 2. API 요청에 필요한 헤더
headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret
}

# ✅ 3. 책 정보 출력 함수
def print_book_info(item):
    print('제목:', item['title'].replace('<b>', '').replace('</b>', ''))  # HTML 태그 제거
    print('저자:', item['author'])
    print('출판사:', item['publisher'])
    print('출판일:', item['pubdate'])
    print('ISBN:', item['isbn'])
    print('링크:', item['link'])
    print('-' * 40)

# ✅ 4. 책 검색 함수
def search_books(query):
    payload = {
        'query': query,
        'display': 30,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/book.json'

    res = requests.get(url, params=payload, headers=headers)

    if res.status_code == 200:
        items = res.json().get('items', [])
        if not items:
            print('검색 결과가 없습니다.')
        for item in items:
            print_book_info(item)
    else:
        print(f'❌ 요청 실패 (상태코드: {res.status_code})')
        print(res.text)

# ✅ 5. 실행 부분
if __name__ == '__main__':
    search_word = input('검색할 책 제목을 입력하세요: ')
    search_books(search_word)
```
