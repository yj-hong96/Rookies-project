```python
import os
import json
import requests

# ✅ 1. 네이버 API 키 입력 (🎯 직접 입력!)
client_id = '8xWBKSelQiATYZJCce6v'
client_secret = 'XUcKjn3UPc'

headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret
}

# ✅ 책 정보 출력 함수
def print_book_info(item):
    print('제목:', item['title'].replace('<b>', '').replace('</b>', ''))
    print('저자:', item['author'])
    print('출판사:', item['publisher'])
    print('출판일:', item['pubdate'])
    print('ISBN:', item['isbn'])
    print('링크:', item['link'])
    print('-' * 40)

# ✅ 책 검색 + JSON 저장 함수
def search_books(query):
    payload = {
        'query': query,
        'display': 30,
        'sort': 'sim'
    }

    url = 'https://openapi.naver.com/v1/search/book.json'

    res = requests.get(url, params=payload, headers=headers)

    if res.status_code == 200:
        items = res.json().get('items', [])

        if not items:
            print('❗ 검색 결과가 없습니다.')
            return

        # ✅ 화면에 출력
        for item in items:
            print_book_info(item)

        # ✅ 2. JSON 파일로 저장 (data/books.json)
        os.makedirs('data', exist_ok=True)
        with open('data/books.json', 'w', encoding='utf-8') as f:
            json.dump(items, f, ensure_ascii=False, indent=2)

        print(f'\n✅ 검색 결과가 data/books.json 에 저장되었습니다.')
    else:
        print(f'❌ 요청 실패 (상태코드: {res.status_code})')
        print(res.text)

# ✅ 실행부
if __name__ == '__main__':
    search_word = input('검색할 책 제목을 입력하세요: ')
    search_books(search_word)
```

```python
import json
import pandas as pd

# 1. JSON 로드
with open('data/books.json', 'r', encoding='utf-8') as f:
    book_data = json.load(f)

# 2. DataFrame 변환
df = pd.DataFrame(book_data)

# 3. 문자열 <b> 태그 제거
df['title'] = df['title'].str.replace('<b>', '', regex=False).str.replace('</b>', '', regex=False)

# 4. 'discount' 숫자로 변환 (정가 대신 사용)
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')

# 5. 2만원 이상 책만 필터링
expensive_books = df[df['discount'] >= 20000]

# 6. 보기 좋게 출력
print("\n💰 2만원 이상 책 목록 (할인가 기준):")
for i, row in expensive_books.iterrows():
    print(f"{i+1}. 제목: {row['title']}")
    print(f"   저자: {row['author']}")
    print(f"   출판사: {row['publisher']}")
    print(f"   발매일: {row['pubdate']}")
    print(f"   할인가: {row['discount']}원")
    print("-" * 50)
```

```python
import json
import pandas as pd

# ✅ 1. JSON 파일 불러오기
with open('data/books.json', 'r', encoding='utf-8') as f:
    book_data = json.load(f)

# ✅ 2. DataFrame으로 변환
df = pd.DataFrame(book_data)

# ✅ 3. 제목에서 <b> 태그 제거
df['title'] = df['title'].str.replace('<b>', '', regex=False).str.replace('</b>', '', regex=False)

# ✅ 4. 출판사가 "인피니티북스"인 책만 필터링
filtered_df = df[df['publisher'].str.contains("인피니티북스", case=False, na=False)]

# ✅ 5. 출력
print("\n🏢 출판사가 '인피니티북스'인 책 목록:")
if filtered_df.empty:
    print("검색된 책이 없습니다.")
else:
    for i, row in filtered_df.iterrows():
        print(f"{i+1}. 제목: {row['title']}")
        print(f"   저자: {row['author']}")
        print(f"   출판사: {row['publisher']}")
        print(f"   발매일: {row['pubdate']}")
        print("-" * 50)
```
